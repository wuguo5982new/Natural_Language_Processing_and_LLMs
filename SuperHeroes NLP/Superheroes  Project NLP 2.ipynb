## Superheroes Project NLP 2

In this part, we found Top10 superheroes with most relatives, and Top10 superheroes with most deceased status.

```python
import numpy as np
import pandas as pd
import ast
## From raw data
df = pd.read_csv('superheroes_nlp_dataset.csv')
# print(df.shape)
```

```python
df.columns
```

```python
df.fillna(value={'history_text':'', 'powers_text':'', 'relatives':''}, inplace=True)   
df['history_text'][df['history_text'] == 'NA'] = ''
df['powers_text'][df['powers_text'] == 'NA'] = ''
df['relatives'][df['relatives'] == 'nan'] = ''

df['history_text'][df['history_text'] == ''] = 'NA'
df['powers_text'][df['powers_text'] == ''] = 'NA'
df['relatives'][df['relatives'] == ''] = 'NA'
df['relatives'][df['relatives'] == 'none'] = 'NA'
df['relatives'][df['relatives'] == 'None'] = 'NA'
```

```python
df['relatives'].head(25).tolist()
```

## Top10 superheroes with most relatives

```python
# tot_teams = df[df['creator'] == creator_unit]['teams'].tolist()
# M = [ast.literal_eval(x) for x in df['relatives'].tolist()]

import collections
family = collections.defaultdict(list)
for i, fam in enumerate(df['relatives'].tolist()):
    if fam not in ['/', 'nan']:
        elements = str(fam).replace(');', ')//').replace('),', ')//').replace(') ;', ')//').replace('\n', '//') \
        .replace(' / ', '//').replace('),\n', ')//').replace(') · ', ')//').replace(') ', ')//').replace('). ', ')//').split('//')
        for unit in elements:
            if unit not in ['NA', 'None', 'Unknown']:
                family[i].append(unit)        
print(family) 
```

```python
## Count the number and sort
top10_relatives = sorted([(i, len(j)) for i,j in dict(family).items()], key=lambda x: x[1], reverse=True)[:10]
top10_relatives
```

```python
indexes = [i for i,j in top10_relatives]
print("Top10 superheroes with most relatives:\n", df.iloc[indexes,]['name'].tolist())
df.iloc[indexes,]
```

## Top10 superheroes with most deceased (status)

```python
dead = []

for i,j in dict(family).items():    
    res = [m.split(',') for m in j]

    for m in res:
        z = len(list([i for i in res if ' deceased)' in i]))
    dead.append((i,z))
# print(dead)

top10_dead_nums = sorted([(i, j) for i,j in dead], key=lambda x: x[1], reverse=True)[:10]
top10_dead_nums
```

```python
indexes_dead = [i for i,j in top10_dead_nums]
print("Top10 superheroes with most decased:\n", df.iloc[indexes_dead,]['name'].tolist())
df.iloc[indexes_dead,]
```
